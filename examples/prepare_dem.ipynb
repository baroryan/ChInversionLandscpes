# Prepare DEM for $\chi_{UKQ}$-Space Inversion

This notebook outlines the basic DEM-preparation step before running the $\chi_{UKQ}$-space inversion.

The goal is to:

1. Download or provide a DEM.
2. Reproject it to a metric coordinate system, preferably UTM.
3. Save the projected DEM for the main inversion notebook.

This step can be done with GDAL, QGIS, Python, or similar GIS tools. Because GDAL can introduce dependency conflicts, this preprocessing can be done outside the main Python environment used for the inversion code.

## 1. Download a DEM

A convenient source for global DEMs is OpenTopography:

https://portal.opentopography.org/datasets

OpenTopography provides access to datasets such as SRTM GL1, SRTM GL3, NASADEM, ALOS World 3D, and Copernicus DEM.

Another option is USGS EarthExplorer:

https://earthexplorer.usgs.gov/

Download a DEM for your study area and save it as:

```text
../data/input_dem.tif
```

## 2. Install GDAL

If needed, create a small separate environment for DEM preparation:

```bash
conda create -n gdal_env -c conda-forge python=3.11 gdal
```

This should be separate from the main inversion environment.

## 3. Reproject DEM to UTM

Use `gdalwarp` to project the DEM to UTM.

To find the appropriate EPSG code for your study area, you can use:

https://epsg.io/

For UTM projections, search by location or UTM zone. For example:

- UTM Zone 11N: `EPSG:32611`
- UTM Zone 11S: `EPSG:32711`

In general, northern-hemisphere UTM zones use `EPSG:326XX`, and southern-hemisphere UTM zones use `EPSG:327XX`, where `XX` is the UTM zone number.

Example for UTM Zone 11N:

```bash
gdalwarp \
  -t_srs EPSG:32611 \
  -r bilinear \
  -of GTiff \
  ../data/input_dem.tif \
  ../data/dem_utm.tif
```

Replace `EPSG:32611` with the appropriate EPSG code for your study area.

## Output

The main inversion notebook expects a projected DEM such as:

```text
../data/dem_utm.tif
```

The DEM should be in a metric coordinate system, preferably UTM.
